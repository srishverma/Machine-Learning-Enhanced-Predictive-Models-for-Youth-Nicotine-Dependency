```python
# Handle missing data using mode imputation
df = df.fillna(df.mode().iloc[0])

# Define the target variable (nicotine addiction prediction)
target_variable = 'EPOUCH'

# Define the predictor variables
predictor_variables = [
    'QN1', 'QN2', 'QN3', 'QN4A', 'AGEGRP', 'SEX', 'SCHOOLTYPE', 'mrace', 'RACE_M', 'RACE_S', 'SEXID', 'SEXID2',
    'ECIGT', 'ECIGAR', 'ESLT', 'EELCIGT', 'EHOOKAH', 'EROLLCIGTS', 'EPIPE', 'ESNUS', 'EDISSOLV', 'EBIDIS', 'EHTP',
    'CCIGT', 'CCIGAR', 'CSLT', 'CELCIGT', 'CHOOKAH', 'CROLLCIGTS', 'CPIPE', 'CSNUS', 'CDISSOLV', 'CBIDIS', 'CHTP',
    'CPOUCH'
]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[predictor_variables], df[target_variable], test_size=0.3, random_state=1)

# Create an SVM classifier
svm_classifier = SVC(kernel='linear', C=1.0)

# Train the SVM classifier on the training data
svm_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred_svm = svm_classifier.predict(X_test)

# Calculate accuracy for the SVM model
accuracy_svm = accuracy_score(y_test, y_pred_svm)

# Display the accuracy of the SVM model
print(f'Accuracy (SVM): {accuracy_svm:.2f}')
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Define the target variable and predictor variables
target_variable = 'EPOUCH'
predictor_variables = ['QN1', 'QN2', 'QN3', 'QN4A', 'AGEGRP', 'SEX', 'SCHOOLTYPE', 'mrace', 'RACE_M', 'RACE_S', 'SEXID', 'SEXID2', 'ECIGT', 'ECIGAR', 'ESLT', 'EELCIGT', 'EHOOKAH', 'EROLLCIGTS', 'EPIPE', 'ESNUS', 'EDISSOLV', 'EBIDIS', 'EHTP', 'CCIGT', 'CCIGAR', 'CSLT', 'CELCIGT', 'CHOOKAH', 'CROLLCIGTS', 'CPIPE', 'CSNUS', 'CDISSOLV', 'CBIDIS', 'CHTP', 'CPOUCH']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[predictor_variables], df[target_variable], test_size=0.3, random_state=1)

# Standardize the feature values
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define a simple feedforward neural network
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(X_train_scaled.shape[1],)),  # Input layer
    tf.keras.layers.Dense(128, activation='relu'),  # Hidden layer with ReLU activation
    tf.keras.layers.Dense(64, activation='relu'),   # Additional hidden layer (optional)
    tf.keras.layers.Dense(1, activation='sigmoid')   # Output layer with sigmoid activation for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the neural network on the training data
model.fit(X_train_scaled, y_train, epochs=10, batch_size=32, verbose=1)

# Make predictions on the test data
y_pred_nn = (model.predict(X_test_scaled) > 0.5).astype(np.int32)

# Calculate accuracy for the neural network model
accuracy_nn = accuracy_score(y_test, y_pred_nn)

# Display the accuracy of the neural network model
print(f'Accuracy (Neural Network): {accuracy_nn:.2f}')
```

In this code:

2. Instances of three different classifiers are created: Support Vector Machine (SVM) with a linear kernel, Decision Tree classifier, and Gaussian Naive Bayes classifier.

3. These classifier instances are stored in a list called `classifiers`.

4. 10-fold cross-validation is performed for each classifier using the `cross_val_score` function. It assesses the model's performance by splitting the data into 10 subsets (folds), training the classifier on 9 of them, and testing it on the remaining fold. This process is repeated 10 times.

6. The results for each classifier, including its class name, mean accuracy, and standard deviation of accuracy, are printed.

This code performs cross-validation for three different classifiers to assess their performance in predicting nicotine addiction. The results provide insights into how well each classifier generalizes to unseen data.

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB


# Define the target variable and predictor variables
target_variable = 'EPOUCH'
predictor_variables = ['QN1', 'QN2', 'QN3', 'QN4A', 'AGEGRP', 'SEX', 'SCHOOLTYPE', 'mrace', 'RACE_M', 'RACE_S', 'SEXID', 'SEXID2', 'ECIGT', 'ECIGAR', 'ESLT', 'EELCIGT', 'EHOOKAH', 'EROLLCIGTS', 'EPIPE', 'ESNUS', 'EDISSOLV', 'EBIDIS', 'EHTP', 'CCIGT', 'CCIGAR', 'CSLT', 'CELCIGT', 'CHOOKAH', 'CROLLCIGTS', 'CPIPE', 'CSNUS', 'CDISSOLV', 'CBIDIS', 'CHTP', 'CPOUCH']

# Create instances of classifiers
svm_classifier = SVC(kernel='linear', C=1.0)
tree_classifier = DecisionTreeClassifier()
nb_classifier = GaussianNB()

# Define a list of classifiers
classifiers = [svm_classifier, tree_classifier, nb_classifier]

# Perform 10-fold cross-validation for each classifier
for clf in classifiers:
    scores = cross_val_score(clf, df[predictor_variables], df[target_variable], cv=10)
    mean_accuracy = scores.mean()
    std_accuracy = scores.std()
    print(f"{clf.__class__.__name__} - Mean Accuracy: {mean_accuracy:.2f}, Std Deviation: {std_accuracy:.2f}")
```

In this code:

1. A strategy for imputing missing values is defined. The specified strategy is 'mean', which means that missing values in the predictor variables (X) will be replaced with the mean of the non-missing values in each column.

2. The SimpleImputer is used to impute the missing values in the predictor variables (X) using the defined strategy.
